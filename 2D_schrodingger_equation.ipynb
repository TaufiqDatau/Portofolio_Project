```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

num_points_x = 50
num_points_y = 50
length_x = 1.0
length_y = 1.0


x = np.linspace(0, length_x, num_points_x)
y = np.linspace(0, length_y, num_points_y)
X, Y = np.meshgrid(x, y)


omega_x = 2 * np.pi
omega_y = np.pi


psi = np.sqrt(2) * np.sin(omega_x * X) * np.sin(omega_y * Y)


psi = psi / np.sqrt(np.trapz(np.trapz(np.abs(psi)**2, x), y))


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.abs(psi)**2, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Probability Density')
ax.set_title('2D Box - Initial Wave Function')
plt.show()
```
